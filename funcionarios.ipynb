## **Passo 3 - Corre√ß√£o do DataFrame (Limpeza, organiza√ß√£o e modelagem)** ##

```python
# Criar um novo dataframe para e manter o originial para diversos EDA, se necess√°rio

# Excluir a coluna ID RH (essa coluna √© desnecess√°ria e n√£o influencia no EDA)
df_analise = df_funcionarios.drop(columns="ID RH")

display(df_analise.head())
```

```python
# Preenceher valores nulos com 0 na coluna Data de Demissao
df_analise.fillna(0, inplace=True)
# Checar o dataframe
display(df_analise.head())

# Conferindo se ainda existem valores ausemtes
display(df_analise.isna().sum())

# Conferir como ficaram os tipos de dados utilizados no dataframe com dtypes 
display(df_analise.dtypes)
```

```python
# Adicionar a coluna Situa√ß√£o para uma an√°lise mais eficaz, uma vez que um dos focos da an√°lise √© a rotatividade 
# dos funcion√°rios, ou seja, as admiss√µes e demiss√µes dos funcion√°rios.

# Adcionar a coluna Situa√ß√£o no dataframe
df_analise["Situa√ß√£o"] = df_analise["Data de Demissao"].apply(lambda x: "Admitido" if x == 0 else "Demitido")

# Visualizar o dataframe com a nova coluna
display(df_analise)
```

Dataframe corrigido e remodelado.

## **Passo 4 - EDA - An√°lise e Visualiza√ß√£o de Dados** ##

---

**An√°lises Descritivas**

```python
# An√°lise da avalia√ß√£o descritiva dos funcion√°rios (cidades X cargos)

# Agrupar por cidade e cargo
print("\nAvalia√ß√£o e quantidade de funcion√°rios por cargo e cidade:")
avaliacao_func = df_analise.groupby(["Cidade", "Cargo"])[["Avalia√ß√£o do Funcion√°rio"]].agg(["mean", "min", "max", "count"]).reset_index().round(2)

# Configura√ß√£o das colunas
avaliacao_func.columns = ["Cidade", "Cargo", "M√©dia da Avalia√ß√£o", "Nota Min√≠ma", "Nota M√°xima", "N¬∫ de Funcion√°rios"]

# Visualizar os dados
display(avaliacao_func)

print(f"""Resumo Geral: 
M√©dia Geral: {avaliacao_func["M√©dia da Avalia√ß√£o"].mean():.2f}
M√©dia Nota M√≠nima: {avaliacao_func["Nota Min√≠ma"].mean():.2f}
M√©dia Nota M√°xima: {avaliacao_func["Nota M√°xima"].mean():.2f}
N¬∫ Total de Funcion√°rios: {avaliacao_func["N¬∫ de Funcion√°rios"].sum()}""")
```

```python
# An√°lise da avalia√ß√£o descritiva dos salarios (cidades X cargos)

# Agrupar por cidade e cargo
print("\nAn√°lise salarial por cargo e cidade:")
analise_salarial = df_analise.groupby(["Cidade", "Cargo"])[["Salario"]].agg(["mean", "min", "max", "sum", "count"]).reset_index().round(2)

# Configura√ß√£o das colunas
analise_salarial.columns = ["Cidade", "Cargo", "M√©dia Salarial", "Valor M√≠nimo", "Valor M√°ximo", "Valor Total", "N¬∫ de Funcion√°rios"]

# Visualizar os dados
display(analise_salarial)

print(f"""Resumo Geral:
M√©dia Salarial Geral: {analise_salarial["Valor Total"].sum() / analise_salarial["N¬∫ de Funcion√°rios"].sum():.2f}
N¬∫ Total de Funcion√°rios: {analise_salarial["N¬∫ de Funcion√°rios"].sum()}
Valor Total dos Sal√°rios: {analise_salarial["Valor Total"].sum():,.2f}""")
```

## **1. Analisar as contrata√ß√µes, funcion√°rios ativos e demitidos, bem como o Turnover (rotatividade).** ##

```python
# Analisar contrata√ß√µes por agrupamento, utilizando o pivot_table

print("\nAn√°lise das contrata√ß√µes:\n")
# An√°lise da situa√ß√£o dos funcion√°rios por rigi√£o
analise_contratacao = df_analise.pivot_table(values="Nome", index="Cidade", columns="Situa√ß√£o", aggfunc="count", fill_value= 0)

# Adiconar a coluna Total
analise_contratacao["Total"] = analise_contratacao.sum(axis=1)

# Adiconar a coluna Percentual % das contrata√ß√µes
analise_contratacao["Percentual %"] = ((analise_contratacao["Admitido"] + analise_contratacao["Demitido"]) / analise_contratacao["Total"].sum()) * 100

# Adiconar a coluna Percentual % turnover
analise_contratacao["Turnover %"] = (analise_contratacao["Demitido"] / analise_contratacao["Total"]) * 100

# Exibir a an√°lise
display(analise_contratacao.round(2).sort_values(by="Admitido", ascending=False))

# Exibir os resultados gerais
print(f"""Resumo das Contrata√ß√µes:
Admitidos: {analise_contratacao["Admitido"].sum()}
Demitidos: {analise_contratacao["Demitido"].sum()}
Total: {analise_contratacao["Total"].sum()}
Turnover: {analise_contratacao["Demitido"].sum() / analise_contratacao["Total"].sum() * 100:.2f}%""")
```

## **2. Explore funcion√°rios por g√™nero e regi√£o.** ##

```python
# # Analisar contrata√ß√µes por g√™nero e regi√£o, agrupando com pivot_table

# An√°lise das contra√ß√µes por g√™nero e regi√£o
print("An√°lise das contrata√ß√µes por g√™nero e regi√£o:")
analise_genero = df_analise.pivot_table(values="Nome", index=["Cidade", "Genero"], columns="Situa√ß√£o", aggfunc="count", fill_value=0)

# Adiconar a coluna Total
analise_genero["Total"] = analise_genero["Admitido"] + analise_genero["Demitido"]

#Adiconar a coluna Percentual %
analise_genero["Percentual Total%"] = analise_genero["Total"] / analise_genero["Total"].sum() * 100

# Visualizando o resultado da an√°lise
display(analise_genero.round(2))

# Pular uma linha
print("\n")

# Criando as vari√°veis do resumo da an√°lise referentes ao funcion√°rios admitidos
total_feminino_admitido = df_analise[df_analise["Genero"] == "F"]["Situa√ß√£o"].value_counts().get("Admitido", 0)

total_masculino_admitido = df_analise[df_analise["Genero"] == "M"]["Situa√ß√£o"].value_counts().get("Admitido", 0)

# Soma total das vari√°veis
total_admitido = total_feminino_admitido + total_masculino_admitido

# Percentual das vari√°veis
perecentual_feminino_admitido = (total_feminino_admitido / total_admitido) * 100

perecentual_masculino_admitido = (total_masculino_admitido / total_admitido) * 100

# Pular uma linha entre o resumo das an√°lises
print("\n")

# Criando as vari√°veis do resumo da an√°lise referentes ao funcion√°rios demitidos
total_feminino_demitido = df_analise[df_analise["Genero"] == "F"]["Situa√ß√£o"].value_counts().get("Demitido", 0)

total_masculino_demitido = df_analise[df_analise["Genero"] == "M"]["Situa√ß√£o"].value_counts().get("Demitido", 0)

# Soma total das vari√°veis
total_demitido = total_feminino_demitido + total_masculino_demitido

# Percentual das vari√°veis
perecentual_feminino_demitido = (total_feminino_demitido / total_demitido) * 100

perecentual_masculino_demitido = (total_masculino_demitido / total_demitido) * 100

# Resultado do resumo das an√°lises:

# Resumo da an√°lise dos admitidos
print(f"""Resumo das admiss√µes por g√™nero e reigi√£o:
Total Feminino Admitido: {total_feminino_admitido} - {perecentual_feminino_admitido:.2f}%
Total Masculino Admitido: {total_masculino_admitido} - {perecentual_masculino_admitido:.2f}%
Total Geral: {total_admitido}""")

print("\n")

# Resumo da an√°lise dos demitidos
print(f"""Resumo das demiss√µes por g√™nero e regi√£o:
Total Feminino Demitido: {total_feminino_demitido} - {perecentual_feminino_demitido:.2f}%
Total Masculino Demitido: {total_masculino_demitido} - {perecentual_masculino_demitido:.2f}%
Total Geral: {total_demitido}""")

print("\n")

# Resumo geral da an√°lise
print(f"""Resumo Geral:
Total Contrata√ß√µes: {total_admitido + total_demitido}
Percentual Masculino: {(total_masculino_admitido + total_masculino_demitido) / (total_admitido + total_demitido) * 100:.2f}%
Percentual Feminino: {(total_feminino_admitido + total_feminino_demitido) / (total_admitido + total_demitido) * 100:.2f}%""")


```

## **3. Investigue como est√£o distribu√≠das as √°reas e cargos entre funcion√°rios e suas respectivas regi√µes.** ##

```python
# An√°lise da distribui√ß√£o dos funicion√°rios entre cargos e √°reas
print("An√°lise da distribui√ß√£o dos funcion√°rios entre √°reas e cargos:")

# An√°lise dos cargos entre √°reas por agrupamento
analise_cargos = df_analise.pivot_table(values="Nome", index=(["√Årea", "Cargo"]), columns="Situa√ß√£o", aggfunc="count", fill_value=0)

# Cri√ß√£o da coluna Total para √°reas e cargos
analise_cargos["Total"] = analise_cargos["Admitido"] + analise_cargos["Demitido"]

# Resultado da an√°lise da distruibui√ß√£o entre cargos e √°reas
display(analise_cargos)

print("\n")

# An√°lise da distribui√ß√£o por √°rea
print("Distribui√ß√£o total por √°rea:")

# An√°lise por √°rea
analise_area = df_analise.pivot_table(values="Nome", index="√Årea", columns="Situa√ß√£o", aggfunc="count", fill_value=0)

# Cria√ß√£o do coluna total para √°rea
analise_area["Total"] = analise_area["Admitido"] + analise_area["Demitido"]

# Resultado da an√°lse por √°rea
display(analise_area)

print("\n")

# An√°lise da distribui√ß√£o por cargos
print("Distribui√ß√£o total por cargos:")

# An√°lise por cargos
analise_cargo = df_analise.pivot_table(values="Nome", index="Cargo", columns="Situa√ß√£o", aggfunc="count", fill_value=0)

# Cria√ß√£o da coluna total para cargos
analise_cargo["Total"] = analise_cargo["Admitido"] + analise_cargo["Demitido"]

# Resultado da an√°lise para cargos
display(analise_cargo)

print("\n")

# Resultado do resumo geral das distribui√ß√µes
print(f"""Resumo Geral:
Total admitidos: {analise_cargos["Admitido"].sum()}
Total Demitidos: {analise_cargos["Demitido"].sum()}
Total Geral: {analise_cargos["Total"].sum()}""")

print("\n")

# An√°lise da distribui√ß√£o dos funcion√°rios entre √°rea e cargos por regi√£o
print("An√°lise da distribui√ß√£o dos funcion√°rios admitidos entre √°rea e cargos por regi√£o:")
analise_regiao = df_analise[df_analise["Situa√ß√£o"] == "Admitido"].pivot_table(values="Nome", index=(["Cidade", "√Årea"]), columns="Cargo", aggfunc="count", fill_value=0)

# Resultado da an√°lise
display(analise_regiao)

```

## **4. Investigue como est√£o distribu√≠dos os cargos por g√™neros.** ##

```python
# An√°lise da distribui√ß√£o dos funicion√°rios entre cargos e g√™neros
print("An√°lise da distribui√ß√£o dos funcion√°rios entre cargos e g√™neros:")

# An√°lise dos cargos entre √°reas e g√™neros por agrupamento
analise_cargo_genero = df_analise.pivot_table(values="Nome", index=(["Cargo", "Genero"]), columns="Situa√ß√£o", aggfunc="count", fill_value=0)

# Cri√ß√£o da coluna Total para cargos e g√™neros
analise_cargo_genero["Total"] = analise_cargo_genero["Admitido"] + analise_cargo_genero["Demitido"]

#Adiconar a coluna Percentual %
analise_cargo_genero["Percentual Total%"] = analise_cargo_genero["Total"] / analise_cargo_genero["Total"].sum() * 100

# Resultado da an√°lise da distruibui√ß√£o entre cargos e g√™neros
display(analise_cargo_genero.round(2))

# Pular uma linha
print("\n")

# Criando as vari√°veis do resumo da an√°lise referentes ao funcion√°rios admitidos
total_feminino_admitido_cargo = df_analise[df_analise["Genero"] == "F"]["Situa√ß√£o"].value_counts().get("Admitido", 0)

total_masculino_admitido_cargo = df_analise[df_analise["Genero"] == "M"]["Situa√ß√£o"].value_counts().get("Admitido", 0)

# Soma total das vari√°veis
total_admitido_cargo = total_feminino_admitido_cargo + total_masculino_admitido_cargo

# Percentual das vari√°veis
perecentual_feminino_admitido_cargo = (total_feminino_admitido_cargo / total_admitido_cargo) * 100

perecentual_masculino_admitido_cargo = (total_masculino_admitido_cargo / total_admitido_cargo) * 100

# Criando as vari√°veis do resumo da an√°lise referentes ao funcion√°rios demitidos
total_feminino_demitido_cargo = df_analise[df_analise["Genero"] == "F"]["Situa√ß√£o"].value_counts().get("Demitido", 0)

total_masculino_demitido_cargo = df_analise[df_analise["Genero"] == "M"]["Situa√ß√£o"].value_counts().get("Demitido", 0)

# Soma total das vari√°veis
total_demitido_cargo = total_feminino_demitido_cargo + total_masculino_demitido_cargo

# Percentual das vari√°veis
perecentual_feminino_demitido_cargo = (total_feminino_demitido_cargo / total_demitido_cargo) * 100

perecentual_masculino_demitido_cargo = (total_masculino_demitido_cargo / total_demitido_cargo) * 100

# Resultado do resumo da an√°lise dos admitidos
print(f"""Resumo das admiss√µes por cargo e g√™nero:
Total Feminino Admitido: {total_feminino_admitido_cargo} - {perecentual_feminino_admitido_cargo:.2f}%
Total Masculino Admitido: {total_masculino_admitido_cargo} - {perecentual_masculino_admitido_cargo:.2f}%
Total Geral: {total_admitido_cargo}""")

# Pular uma linha entre o resumo das an√°lises
print("\n")

# Visulizar o resumo da an√°lise dos demitidos
print(f"""Resumo das demiss√µes por cargo e g√™nero:
Total Feminino Demitido: {total_feminino_demitido_cargo} - {perecentual_feminino_demitido:.2f}%
Total Masculino Demitido: {total_masculino_demitido_cargo} - {perecentual_masculino_demitido_cargo:.2f}%
Total Geral: {total_demitido_cargo}""")

print("\n")

print(f"""Resumo Geral:
Total Contrata√ß√µes: {total_admitido_cargo + total_demitido_cargo}
Percentual Masculino: {(total_masculino_admitido_cargo + total_masculino_demitido_cargo) / (total_admitido_cargo + total_demitido_cargo) * 100:.2f}%
Percentual Feminino: {(total_feminino_admitido_cargo + total_feminino_demitido_cargo) / (total_admitido_cargo + total_demitido_cargo) * 100:.2f}%""")

```

## **5. Identifique a distribui√ß√£o de sal√°rios entre funcion√°rios e seus respectivos cargos, como tamb√©m as horas extras praticadas em cada cargo.** ##
